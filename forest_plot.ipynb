```python
# Importar as bibliotecas necessárias
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
import zepid
from zepid.graphics import EffectMeasurePlot

# Substitua os valores de exemplo pelos seus próprios dados

# Replace the example values ​​with your own data
labs = ["Label1 (Isq=41.37% Tausq=0.146 pvalue=0.039 )",
        "Label2 (Isq=25.75% Tausq=0.092 pvalue=0.16 )",
        "Label3 (Isq=60.34% Tausq=0.121 pvalue=0.00 )",
        "Label4 (Isq=25.94% Tausq=0.083 pvalue=0.16 )",
        "Label5 (Isq=74.22% Tausq=0.465 pvalue=0.00 )"]
measure = [2.09, 2.24, 1.79, 2.71, 1.97]
lower = [1.49, 1.63, 1.33, 2.00, 1.25]
upper = [2.92, 3.07, 2.42, 3.66, 3.11]

# Criar uma instância do gráfico de EffectMeasurePlot
p = EffectMeasurePlot(label=labs, effect_measure=measure, lcl=lower, ucl=upper)

# Configurações opcionais para personalização
p.labels(effectmeasure='RR')  # Escolher rótulo para a medida de efeito (pode ser 'RR', 'OR', 'RD', etc.)
p.colors(pointshape="D")       # Escolher forma dos pontos nos gráficos

# Criar o gráfico
ax = p.plot(figsize=(7, 3), t_adjuster=0.09, max_value=4, min_value=0.35)

# Configurações adicionais de título, rótulos, bordas, etc.
plt.title("Modelo de Efeito Aleatório (Razão de Risco)", loc="right", x=1, y=1.045)
plt.suptitle("Método de Imputação de Dados Ausentes", x=-0.1, y=0.98)
ax.set_xlabel("Favorece Controle       Favorece Haloperidol", fontsize=10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(False)

# Salvar o gráfico como uma imagem
plt.savefig("MetodoImputacaoDadosAusentes.png", bbox_inches='tight')
```
